# DATA PREPROCESSING

### NULL VALUES

```python
# check null values
df.isnull().sum()

# remove null values (if any)
df = df.dropna()
```

### DUPLICATE VALUES

```python
# check duplicates
df.duplicated().sum()

# remove duplicates
df = df.drop_duplicates()
```

### UNNAMED COLUMNS

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

### DATA CONVERSTION

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Brand'] = le.fit_transform(df['Brand'])
```

# FEATURE TRANSFORMATION

```python
# example new feature
df['Performance_Index'] = df['Processor_Speed'] * df['RAM_Size']
```

# FEATURE SELECTION

```python
X = df.drop('Price', axis=1)   # input features
y = df['Price']                # output
```

# FEATURE SCALING

### STANDERDIZATION

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### NORMALIZATION

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

# SPLITTING DATA

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

# MODEL SELECTION

```python
# Import required libraries
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score

# Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)

# SVM Model
svm = SVR()
svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)

# Decision Tree Model
dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)

# Random Forest Model
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)

# Polynomial Regression
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)
poly_pred = poly_model.predict(X_test_poly)

# Print R2 Scores of all models
print("Linear Regression R2 Score:", r2_score(y_test, lr_pred))
print("SVM R2 Score:", r2_score(y_test, svm_pred))
print("Decision Tree R2 Score:", r2_score(y_test, dt_pred))
print("Random Forest R2 Score:", r2_score(y_test, rf_pred))
print("Polynomial Regression R2 Score:", r2_score(y_test, poly_pred))
```

### model prediction

```python
import pandas as pd

y_pred = lr.predict(X_test)

results = pd.DataFrame({
    "Actual_Price": y_test,
    "Predicted_Price": y_pred
})

print(results.head(10))
```

```python
import pickle
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)

with open("model.pkl", "wb") as f:
    pickle.dump(model, f)

with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

print("Model and scaler saved successfully")
```
